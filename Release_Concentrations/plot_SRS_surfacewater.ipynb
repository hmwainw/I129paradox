```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
m2r = 0.176550591 #piCi/ng, mass to Ci
```

```python
d1 = pd.read_csv('FM-6.csv')
date1 = pd.to_datetime(d1['SAMPLE_DATE'])
y1 = d1['REPORT_RESULT_VALUE']/m2r
max(y1)
```

```python
np.mean(y1)
```

```python
y1.shape
```

```python
d2 = pd.read_csv('FMC-002F.csv')
date2 = pd.to_datetime(d2['SAMPLE_DATE'])
y2 = d2['REPORT_RESULT_VALUE']/m2r
max(y2)
```

```python
np.mean(y2)
```

```python
y2.shape
```

```python
plt.figure(figsize=(6, 3))
plt.plot(date2,y2,label='SRS F-Area')
plt.plot(date1,y1,label='Formile Creek')
plt.ylabel('I-129 concentrations, ng/L')
plt.legend()
```
